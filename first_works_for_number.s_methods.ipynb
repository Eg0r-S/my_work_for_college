## Упражнение 2

Уточните наименьший по модулю отличного от нуля корень уравнения xsinx - 1 = 0 методом простой итерации с толщиной до 10^-5, используя расчётную таблицу и калькулятор, программу на ПК.

```python
def f(x): return x * np.sin(x) - 1


# Диапазон x
x = np.linspace(-10, 10, 1000)


plt.figure(figsize=(10, 6))
plt.plot(x, f(x), 'b-', linewidth=2, label='f(x) = x*sin(x) - 1')
plt.axhline(y=0, color='k', linestyle='--', alpha=0.7)
plt.axvline(x=0, color='k', linestyle='--', alpha=0.7)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции f(x) = x*sin(x) - 1')
plt.tight_layout()
plt.show()
```

Функция пересекает ось х в областях точек: -9, -6, -2.7, -1, 0, 1, 2.7, 6, 9 по оси х. Дальше я беру точку x = 1, как начальное приближение для метода простой итерации.

```python
# Итерационная функция от x * Sin(x) - 1 = 0
def iterarion_f(x): return 1 / np.sin(x)

def simple_iteration(x0, eps):
    table_data = []

    x_prev = x0
    x_new = iterarion_f(x_prev)
    iter_count = 1

    table_data.append({
            'Итерация': iter_count,
            'x_n': x_prev,
            'x_{n+1}': x_new,
            '|Δx|': abs(x_new - x_prev)
        })
    
    while abs(x_new - x_prev) > eps:
        x_prev = x_new
        x_new = iterarion_f(x_prev)
        iter_count += 1

        table_data.append({
            'Итерация': iter_count,
            'x_n': x_prev,
            'x_{n+1}': x_new,
            '|Δx|': abs(x_new - x_prev)
        })
    
    df = pd.DataFrame(table_data)
    df = df.round({
        'x_n': 6,
        'x_{n+1}': 6,
        '|Δx|': 8
    })

    return round(x_new, 5), df

# Изначальные значения
eps = 1e-5
x0 = 1


root, df = simple_iteration(x0, eps)


print(df.to_string(index=False))
print(f"\nНайденный наименьший по модулю корень уравнения: x ≈ {root}")
```

## Упражнение 2:

Комбинационным методом хорд и касательных уточните до 0.000001 корень уравнения x^3 + 3x^2 - 24x + 1 = 0 на отрезке [0;1]

```python
def f(x):
    return x**3 - 3 * x**2 - 24 * x + 1     # Функция

def df(x):
    return 3 * x**2 - 6 * x - 24  # Производная функция

def df2(x):
    return 6 * x - 6  # Вторая производная функции

def solve_combined(a, b, eps):

    # Определяю, с какой стороны касательная (x_k), а с какой хорда (x_n)
    if f(a) * df2(a) > 0:
        x_k, x_n = a, b
    else:
        x_k, x_n = b, a
    
    # Списки для таблицы
    it = []
    x_ns = []
    x_ks = []

    iteration = 0
    while abs(x_n - x_k) > eps:

        # Временно создаю новые переменные, чтобы изменения не мешали друг другу.
        x_n_new = x_n - (f(x_n) * (x_k - x_n)) / (f(x_k) - f(x_n))
        x_k_new = x_k - f(x_k) / df(x_k)
        
        x_n, x_k = x_n_new, x_k_new


        # Добавляю данные в списки
        x_ns.append(x_n)
        x_ks.append(x_k)
        iteration += 1
        it.append(iteration)
    

    # Создаю таблицу
    data = {'Итерации': it, 'x_n': x_ns, 'x_k': x_ks}
    dff = pd.DataFrame(data)
    dff = dff.set_index('Итерации')
    print(dff)

    root = (x_n + x_k) / 2
    print(f'\nНайденный корень уравнения: {round(float(root), 6)}')


# Изначальные параметры
a = 0
b = 1
eps = 1e-6  # Точность

solve_combined(a, b, eps)
```

## №1

Найдите методом половинного деления корень трансцедентного уравнения lg x - cos x = 0 с четырьмя знаками после запятой в MathCad. Корни отделите графически.

Проверка через программу на Python:

```python
# Первая функция
def f1(x):
    return np.log10(x)


# Вторая функция
def f2(x):
    return np.cos(x)


# Если f_diff = 0, то корень найден.
def f_diff(x):
    return f1(x) - f2(x)


# Диапазон x.
x_plot = np.linspace(-6.0, 10, 1000)


# Графики функций
plt.figure(figsize=(8, 5))
plt.plot(x_plot, f1(x_plot), label=r'$y = \ln(x + 6.1)$', color='blue', linewidth=2)
plt.plot(x_plot, f2(x_plot), label=r'$y = 2\sin(x - 1.4)$', color='red', linestyle='--')


# Оформление графика
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle=':', alpha=0.7)
plt.title('Графический метод отделения корней')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

Графики совпадают.

Исходя из графиков, функции пересекаются в областях точек: 1.5, 5.7, 6.9. Далее я буду искать корень в диапозоне [1;2]

Проверка с помощью моей программы на Python

```python
# Определяю функцию f(x) = lg(X) - Cos(x)
def f(x):
    return np.log10(x) - np.cos(x)

def bisection_method(a, b, eps):

    # Проверка условия теоремы Больцано-Коши
    if f(a) * f(b) > 0:
        print("На данном отрезке корней нет или их четное количество")
        return None
    

    # Списки для таблицы
    it = []
    ass = []
    bs = []
    result = []

    
    iteration = 0
    while (b - a) / 2 > eps:
        
        mid = (a + b) / 2


        # Выбираю половину отрезка, на которой функция меняет знак
        if f(a) * f(mid) < 0:
            b = mid
        else:
            a = mid
        

        # Добавляю в списки данные
        ass.append(round(a, 4))
        bs.append(round(b, 4))
        result.append(round(mid, 4))
        iteration += 1
        it.append(iteration)
    

    # Создание таблицы
    data = {'Итерации': it, 'a': ass, 'b': bs, 'result': result}
    df = pd.DataFrame(data)
    df = df.set_index('Итерации')

    return df, (a + b) / 2

a, b = 1, 2
epsilon = 1e-4 # Точность

df, root = bisection_method(a, b, epsilon)

if root is not None:
    print(df)
    print(f"\nКорень уравнения: {root:.4f}")
```

Корни совпали в результатах программы Calc и моего кода. Это подтверждает коректность вычислений.

## №3

Найдите наибольший положительный корень уравнения 4x-3lnx=4 с точностью 10^-4, используя метод итераций в системе Derive. Корни отделите графически.

Функции пересекаются только в одной точке x = 1. Я беру начальное приближение x = 2, потому что при х = 1 программа не выводит корень с точностью до 0,0001

ПРоверка:

```python
def phi(x):

    # Преобразованная функция f(x) = 4х - 3ln(x) = 4
    return 1 + 0.75 * np.log(x)

def simple_iteration(x0, eps):


    # Списки для данных
    it = []
    x_olds = []
    x_news = []
    
    x_old = x0
    x_new = phi(x_old)
    iterations = 1

    it.append(iterations)
    x_olds.append(x_old)
    x_news.append(x_new)

    while abs(x_new - x_old) > eps:
        x_old = x_new
        x_new = phi(x_old)
        iterations += 1


        # Добавление данных в списки
        it.append(iterations)
        x_olds.append(x_old)
        x_news.append(x_new)
    
    # Создание таблицы
    data = {'Итерации': it, 'x_old': x_olds, 'x_new': x_news}
    df = pd.DataFrame(data)
    df = df.set_index('Итерации')
    return df, x_new

# Начальные параметры
x_start = 2  
precision = 1e-4    # Точность

simple_iteration(x_start, precision)
df, root = simple_iteration(x_start, precision)

print(df)
print(f"\nКорень уравнения: {root:.4f}")
```

Результаты вычислений совпали. Это подтверждает коректность расчётов.
