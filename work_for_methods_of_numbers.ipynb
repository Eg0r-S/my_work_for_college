## Упражнение 2

Уточните наименьший по модулю отличного от нуля корень уравнения xsinx - 1 = 0 методом простой итерации с толщиной до 10^-5, используя расчётную таблицу и калькулятор, программу на ПК.

```python
def f(x): return x * np.sin(x) - 1


x = np.linspace(-10, 10, 1000)


plt.figure(figsize=(10, 6))
plt.plot(x, f(x), 'b-', linewidth=2, label='f(x) = x*sin(x) - 1')
plt.axhline(y=0, color='k', linestyle='--', alpha=0.7)
plt.axvline(x=0, color='k', linestyle='--', alpha=0.7)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции f(x) = x*sin(x) - 1')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def iterarion_f(x): return 1 / np.sin(x)    #Итерационная функция от x * Sin(x) - 1 = 0

def simple_iteration(x0, eps):
    table_data = []

    x_prev = x0
    x_new = iterarion_f(x_prev)
    iter_count = 1

    table_data.append({
            'Итерация': iter_count,
            'x_n': x_prev,
            'x_{n+1}': x_new,
            '|Δx|': abs(x_new - x_prev)
        })
    
    while abs(x_new - x_prev) > eps:
        x_prev = x_new
        x_new = iterarion_f(x_prev)
        iter_count += 1

        table_data.append({
            'Итерация': iter_count,
            'x_n': x_prev,
            'x_{n+1}': x_new,
            '|Δx|': abs(x_new - x_prev)
        })
    
    df = pd.DataFrame(table_data)
    df = df.round({
        'x_n': 6,
        'x_{n+1}': 6,
        '|Δx|': 8
    })

    return round(x_new, 5), df

# Изначальные значения
eps = 1e-5
x0 = 1.8


root, df = simple_iteration(x0, eps)


print(df.to_string(index=False))
print(f"\nНайденный наименьший по модулю корень уравнения: x ≈ {root}")
```

Сопоставляя рис. 2.10, 2.11 и 2.12, проиллюстрируйте графически решение уравнения методами хорд и касательных для всех четырех ситуаций, изображенных на рис. 2.9.

```python
# Функция 1: парабола вверх (F'>0, F''>0)
def f(x): return (x - 1.5)**2 - 0.5
root = 1.5 + np.sqrt(0.5)

x = np.linspace(0.5, 2.8, 200)
y = f(x)

plt.figure(figsize=(5, 5))
plt.title('Случай a): F\' > 0, F\'\' > 0', fontsize=12, fontweight='bold')

plt.plot(x, y, 'b-', linewidth=2.5, label='f(x)')
plt.scatter(root, f(root), color='black', zorder=5, label='Корень')
plt.grid(True, linestyle=':', alpha=0.7)

plt.tight_layout()
plt.show()

print(f'Корень = {root:.2f}')
```

```python
# Функция 2: парабола вниз (F'<0, F''<0)
f = lambda x: -(x - 1.5)**2 + 0.5
root = 1.5 - np.sqrt(0.5)

x = np.linspace(0.5, 2.8, 200)
y = f(x)

plt.figure(figsize=(5, 5))
plt.plot(x, y, 'b-', linewidth=2.5, label='f(x)')
plt.scatter(root, f(root), color='black', zorder=5, label='Корень')

plt.title('Случай b): F\' < 0, F\'\' < 0', fontsize=12, fontweight='bold')
plt.grid(True, linestyle=':', alpha=0.7)

plt.tight_layout()
plt.show()

print(f'Корень = {root:.2f}')
```

```python
# Функция 3: убывающая, вогнутая вверх (F'<0, F''>0)
def f(x): return -np.sqrt(x - 1) + 0.5
root = 1.25

x = np.linspace(1.0, 2.8, 200)  # область определения от 1
y = f(x)

plt.figure(figsize=(5, 5))
plt.plot(x, y, 'b-', linewidth=2.5, label='f(x)')
plt.scatter(root, f(root), color='black', zorder=5, label='Корень')

plt.title('Случай c): F\' < 0, F\'\' > 0', fontsize=12, fontweight='bold')
plt.grid(True, linestyle=':', alpha=0.7)

plt.tight_layout()
plt.show()

print(f'Корень = {root:.2f}')
```

## Упражнение 2:

Комбинационным методом хорд и касательных уточните до 0.000001 корень уравнения x^3 + 3x^2 - 24x + 1 = 0 на отрезке [0;1]

```python
def f(x):
    return x**3 - 3 * x**2 - 24 * x + 1

def df(x):
    return 3 * x**2 - 6 * x - 24  # Производная функция

def df2(x):
    return 6 * x - 6  # Вторая производная функции

def solve_combined(a, b, eps):

    # Определяю, с какой стороны касательная (x_k), а с какой хорда (x_n)
    if f(a) * df2(a) > 0:
        x_k, x_n = a, b
    else:
        x_k, x_n = b, a
    
    # Списки для таблицы
    it = []
    x_ns = []
    x_ks = []

    iteration = 0
    while abs(x_n - x_k) > eps:

        # Временно создаю новые переменные, чтобы изменения не мешали друг другу.
        x_n_new = x_n - (f(x_n) * (x_k - x_n)) / (f(x_k) - f(x_n))
        x_k_new = x_k - f(x_k) / df(x_k)
        
        x_n, x_k = x_n_new, x_k_new


        # Добавляю данные в списки
        x_ns.append(x_n)
        x_ks.append(x_k)
        iteration += 1
        it.append(iteration)
    

    # Создаю таблицу
    data = {'Итерации': it, 'x_n': x_ns, 'x_k': x_ks}
    dff = pd.DataFrame(data)
    dff = dff.set_index('Итерации')
    print(dff)

    root = (x_n + x_k) / 2
    print(f'\n{round(float(root), 6)}')


# Изначальные параметры
a = 0
b = 1
eps = 1e-6  # Точность

solve_combined(a, b, eps)
```
