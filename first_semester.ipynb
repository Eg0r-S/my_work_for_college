Циклы (Вариант 24)

Работа 1

```python
r = float(input("Введите радиус R (R > 0): ")) # 1
count_points = 0
if r <= 0:
    print("Радиус не подходит условию")
else:
    
    for x in range(-round(r), round(r)+1):
        for y in range(-round(r), round(r)+1):
            if x*x + y*y <= r*r:
                count_points += 1
    print(f'Количество точек с целочисленными координатами = {count_points}')
```

Работа 3

```python

epsilon = 0.0001  # погрешность
a_list = []  # Список последовательности
a_1 = 1.0  # Первое число последовательности, которое равно 1.
a_list.append(a_1)
a_n = a_1  # Переменная для работы цикла, в которой перебираются значения последовательности
n = 1
while True:
    n += 1
    a_n_minus_1 = a_n
    a_n = 0.5 * (a_n_minus_1 + 2 / a_n_minus_1)  # 1/2(a_n-1 + 2/a_n-1), a_1 = 1, M: |a_n^2 - 2| < ε
    a_list.append(a_n)
    if abs(a_n ** 2 - 2) < epsilon:
        print(f"Наименьший номер элемента n: {n}")
        print("Все элементы от а_1 до а_n:")
        for i, a_i in enumerate(
                a_list):  # Функция enumerate() - считывает список и возвращает элемент с его индексом (индекс элемента, элемент).
            print(f"a_{i+1} = {a_i}")
        break
else:
    print('В последовательности нет числа, которое удовлетворяет условию.')
```

Работа 4 (библиотека для таблицы)

```python
import pandas as pd
import math

# Определение диапазона и шага
start_val, end_val = map(float, input("Введите диапозон в формате 'a b':").split())  # -0.5 0.5            10 1
step = float(input("Введите шаг для диапозона: "))  # 0.1                                                    0

element = start_val

x_values = []
f_values = []

if start_val > end_val and step >= 0:
    print("Неверный ввод!")
else:
    while True:
        if element > end_val:
            break

        x_values.append(element)

        if step == 0:
            break
        element += step

    # Вычисление значений функции F(x) = tg(2x) - 3
    for x in x_values:
        f_values.append(math.tan(2 * x) - 3)

    # Округление значений аргумента и функции
    for i in range(len(x_values)):
        x_values[i] = round(x_values[i], 2)

    for i in range(len(f_values)):
        f_values[i] = round(f_values[i], 2)

    # Создание DataFrame (таблицы) с использованием pandas
    data = {'Результат': " ", 'Значение аргумента (x)': x_values, 'Значение функции F(x)': f_values}
    df = pd.DataFrame(data)
    df = df.set_index('Результат')
    # Вывод таблицы на экран
    df
```

```python
n1, n2 = map(int, input("Введите первое и последнее число последовательности целых чисел ").split())
count = 0

if n1 > n2:
    print("Неверный ввод.")
else:
    for i in range(n1, n2+1):
        if i < 0 and abs(i)%2!=0:
            count += 1
    if count == 0:
        print("В последовательности нет целых отрицательных нечётных чисел")
    else:
        print(f"В последовательности от {n1} до {n2} содержится {count} нечётных отрицательных чисел.")
```
