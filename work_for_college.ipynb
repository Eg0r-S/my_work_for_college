```python
import math

def solve_iteration(x0, eps):
    print(f"{'n':<5} | {'x_n':<10} | {'|x_n - x_{n-1}|':<15}")
    print("-" * 35)
    
    x_prev = x0
    n = 0
    
    while True:
        n += 1
        # Итерационная формула: x = 1 / sin(x)
        try:
            x_curr = 1 / math.sin(x_prev)
        except ZeroDivisionError:
            print("Ошибка: деление на ноль (sin(x) = 0)")
            return None
            
        diff = abs(x_curr - x_prev)
        print(f"{n:<5} | {x_curr:<10.6f} | {diff:<15.6f}")
        
        if diff < eps:
            return x_curr
        
        x_prev = x_curr
        
        # Защита от бесконечного цикла
        if n > 1000:
            print("Превышено максимальное число итераций")
            return None

# Параметры задачи
epsilon = 1e-5
initial_guess = 1.1  # Начальное приближение вблизи корня

root = solve_iteration(initial_guess, epsilon)

if root:
    print("-" * 35)
    print(f"Найденный корень: {root:.5f}")
    print(f"Проверка: f(x) = {root * math.sin(root) - 1:.7f}")
```
